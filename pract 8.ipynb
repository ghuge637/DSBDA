```python
import seaborn as sns
df = sns.load_dataset('titanic')
df
```

```python
sns.distplot(x = df['age'], bins = 10)
```

```python
sns.distplot(x = df['age'], bins = 10, kde = False)
```

```python
#jointplot - combination of 2 distplot

sns.jointplot(x = df['age'], y = df['fare'], kind = 'scatter')

#plot 1
```

```python
sns.jointplot(x = df['age'], y = df['fare'], kind = 'hex')
```

```python
#categorical plot
#bar plot
sns.barplot(x = df['sex'], y = df['age'], data = df)
```

```python
#count plot
sns.countplot(x= df['sex'], data=df)
```

```python
#boxplot sjows categorical data into quartile data
sns.boxplot(x=df['sex'], y=df['age'], data = df)

# value below the box is first quartile 
```

```python
sns.boxplot(x=df['sex'], y=df['age'] , data = df, hue = 'survived')
```

```python
#violinplot

sns.violinplot(x=df['sex'], y=df['age'], data=df)
```

```python
sns.violinplot(x=df['sex'], y=df['age'], data=df, hue = 'survived')
```

```python
sns.stripplot(x='sex', y='age', data=df)
```

```python
sns.swarmplot(x='sex', y='age', data=df)
```

```python
sns.swarmplot(x='sex', y='age', data=df, hue = 'survived')
```
